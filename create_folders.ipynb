# Create Folder Structure for Python AI/ML Repository

This notebook creates the folder structure for organizing Python, AI, ML, Vision, Deep Learning, and GenAI content.

```python
# Import Required Modules
import os
```

```python
# Define Main Folder Names
main_folders = ['python', 'ai', 'ml', 'vision', 'deeplearning', 'genai']
```

```python
# Create Main Folders
for folder in main_folders:
    os.makedirs(folder, exist_ok=True)
    print(f"Created main folder: {folder}")
```

```python
# Define Subfolder Names
subfolders = {
    'python': ['basics', 'libraries', 'projects'],
    'ai': ['fundamentals', 'algorithms'],
    'ml': ['supervised', 'unsupervised', 'ensemble'],
    'vision': ['computer-vision', 'image-processing'],
    'deeplearning': ['neural-networks', 'cnn', 'rnn', 'transformers'],
    'genai': ['llms', 'prompt-engineering', 'fine-tuning']
}
```

```python
# Create Subfolders
for main_folder, subs in subfolders.items():
    for sub in subs:
        path = os.path.join(main_folder, sub)
        os.makedirs(path, exist_ok=True)
        print(f"Created subfolder: {path}")
```

```python
# Verify Folder Structure
print("Folder structure created:")
for root, dirs, files in os.walk('.'):
    level = root.replace('.', '').count(os.sep)
    indent = ' ' * 2 * level
    print(f"{indent}{os.path.basename(root)}/")
    subindent = ' ' * 2 * (level + 1)
    for d in dirs:
        print(f"{subindent}{d}/")
```
