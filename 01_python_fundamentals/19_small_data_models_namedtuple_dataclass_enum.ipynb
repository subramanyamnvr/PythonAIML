## What changes in memory while this code runs

These tools still create ordinary Python objects, but they automate common patterns so you do not write unnecessary boilerplate. The runtime object model stays familiar even when the syntax gets nicer.


```python
from collections import namedtuple
from dataclasses import dataclass
from enum import Enum, auto

Point = namedtuple("Point", "x y")
@dataclass
class User:
    name: str
    age: int
class Status(Enum):
    PENDING = auto()
    DONE = auto()

print(Point(1, 2))
print(User("Ada", 31))
print(Status.PENDING)
```

## What compiled bytecode reveals about this code

The bytecode is not the main story here, but it is still helpful to remember that these conveniences end up producing normal class-like runtime behavior that the interpreter executes in ordinary ways.


```python
import dis
from dataclasses import dataclass

@dataclass
class Person:
    name: str
    age: int

def build():
    return Person("Ada", 31)

dis.dis(build)
```

## namedtuple for a small point record

You get named access without a custom full class body.


```python
from collections import namedtuple
Point = namedtuple("Point", "x y")
p = Point(3, 4)
print(p.x, p.y)
```

## dataclass for simple domain data

This is often the sweet spot for many application models.


```python
from dataclasses import dataclass

@dataclass
class Book:
    title: str
    year: int

print(Book("Deep Python", 2026))
```

## Enum for fixed choices

Enums reduce typo-prone “stringly typed” status values.


```python
from enum import Enum

class Mode(Enum):
    TRAIN = "train"
    TEST = "test"

print(Mode.TRAIN)
```
