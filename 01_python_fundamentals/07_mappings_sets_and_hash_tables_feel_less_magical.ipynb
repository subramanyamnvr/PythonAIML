## What changes in memory while this code runs

A dictionary is not just a list of pairs searched one by one. Internally it uses a hash-based organization so Python can usually jump near the right location quickly. The memory layout is more complex than a beginner needs to memorize, but the key idea is fast lookup through hashing.


```python
d = {"name": "Ada", "role": "Engineer"}
s = {1, 2, 3}
print(id(d), d)
print(id(s), s)
```

## What compiled bytecode reveals about this code

Bytecode shows dictionary and set construction instructions and the operations Python uses for subscription and membership checks. It is a nice reminder that these are specialized runtime operations.


```python
import dis

def lookup(data, key):
    return data[key]

dis.dis(lookup)
```

## Dictionary creation and safe lookup

Use `get` when absence is normal rather than exceptional.


```python
person = {"name": "Ada", "role": "Engineer"}
print(person.get("name"))
print(person.get("age", "unknown"))
```

## Set membership and deduplication

Sets answer “have I seen this?” very well.


```python
values = [1, 2, 2, 3, 3, 3]
unique = set(values)
print(unique)
print(2 in unique)
```

## A list cannot be a dictionary key

Lists are mutable, so they are unhashable.


```python
try:
    data = {[1, 2]: "bad key"}
except TypeError as exc:
    print(exc)
```
