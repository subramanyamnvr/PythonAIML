A dictionary is not just a list of pairs searched one by one. Internally it uses a hash-based organization so Python can usually jump near the right location quickly. The memory layout is more complex than a beginner needs to memorize, but the key idea is fast lookup through hashing.


```python
d = {"name": "Ada", "role": "Engineer"}
s = {1, 2, 3}
print(id(d), d)
print(id(s), s)
```

Bytecode shows dictionary and set construction instructions and the operations Python uses for subscription and membership checks. It is a nice reminder that these are specialized runtime operations.


```python
import dis

def lookup(data, key):
    return data[key]

dis.dis(lookup)
```

A deeper mapping/set model is really a deeper hash model. Dictionaries and sets feel powerful because they turn many questions into fast lookup questions, but they only stay reliable if equality and hashing agree. That contract is one of the most important invisible rules in Python.


```python
class Key:
    def __init__(self, value):
        self.value = value
    def __eq__(self, other):
        return isinstance(other, Key) and self.value == other.value
    def __hash__(self):
        return hash(self.value)

a = Key('x')
b = Key('x')
print(a == b)
print(hash(a), hash(b))
print({a: 'works'}[b])
```

The bigger lesson in this notebook is that data structures are design decisions, not only storage choices. A dictionary says ?lookup by key matters here?. A set says ?uniqueness or membership matters here?. Those choices affect performance, readability, and even bug patterns. When you understand hashing properly, these tools stop feeling like convenient black boxes and start feeling like deliberate engineering instruments.
