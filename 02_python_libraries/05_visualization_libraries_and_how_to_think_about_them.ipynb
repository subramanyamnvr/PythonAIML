# Visualization Libraries and How to Think About Them
Plotting libraries are often learned as a pile of function calls, but the deeper skill is knowing what kind of question each chart answers and how the tooling around that chart fits your workflow. Even when a package is missing locally, that mental model is still worth building carefully.

This environment does not currently have matplotlib, seaborn, plotly, or streamlit installed, so the notebook is written to execute safely while still teaching the conceptual and practical differences among them.

```python
import importlib.util
libs = ['matplotlib', 'seaborn', 'plotly', 'streamlit']
for name in libs:
    print(name, 'installed' if importlib.util.find_spec(name) else 'not installed')
```

Matplotlib is the foundational plotting library that gives detailed control over figures and axes. Seaborn sits on top of it and makes statistical plotting patterns easier. Plotly emphasizes interactive visuals. Streamlit is more about wrapping analysis and visuals into a lightweight app experience.

```python
chart_selection = {
    'distribution': ['histogram', 'box plot', 'density plot'],
    'comparison': ['bar chart', 'grouped bar chart'],
    'relationship': ['scatter plot', 'line chart'],
    'composition': ['stacked bar', 'area chart'],
}
for question, charts in chart_selection.items():
    print(question, '->', ', '.join(charts))
```

A helpful way to study visualization is to separate the tool from the analytical job. A scatter plot asks about relationship, a histogram asks about distribution, a line chart asks about change over ordered time, and a heatmap asks about intensity across a grid.

```python
sample_guidelines = [
    'Always label axes with meaning, not only variable names.',
    'Choose a baseline carefully for bar charts.',
    'Use color to encode meaning, not decoration alone.',
    'Reduce chart junk before adding annotation.',
    'Prefer one clear story over ten competing encodings.'
]
for item in sample_guidelines:
    print('-', item)
```

For interviews, a strong answer about visualization libraries often focuses less on memorizing every API call and more on choosing the right visual grammar, avoiding misleading encodings, and knowing which library fits static reports, statistical EDA, interactive dashboards, or quick app prototypes.

```python
toy_data = [12, 14, 18, 18, 19, 22, 25, 25, 25, 31]
print('min:', min(toy_data))
print('max:', max(toy_data))
print('mean:', sum(toy_data) / len(toy_data))
print('sorted:', sorted(toy_data))
```

Even without plotting, you can practice the chart-thinking habit by deciding which chart you would use, what pattern you expect to see, and what misleading design choice you would avoid. That kind of reasoning transfers well when the actual plotting library is available.

## Final Takeaway
The deepest visualization skill is not tied to one package. It is the ability to match a question to a visual form, communicate honestly, and choose a library whose strengths fit the workflow. APIs change; clear thinking about visual meaning lasts much longer.
