# Vectors, Geometry, and Similarity
A vector can be read as a list of numbers, a point in space, a direction, or a bundle of features. ML becomes easier when those interpretations stop feeling separate and start feeling like different camera angles on the same object.

Vectors show up everywhere in ML because data points, embeddings, gradients, and parameter collections can all be represented as vectors. The notation may look abstract, but the underlying idea is often just a structured list of related numerical values.

```python
import numpy as np

v = np.array([3.0, 4.0])
w = np.array([1.0, 2.0])
print('v:', v)
print('w:', w)
print('v + w:', v + w)
print('2 * v:', 2 * v)
print('norm of v:', np.linalg.norm(v))
```

Length, often called norm or magnitude, tells you how large a vector is. In geometry this is distance from the origin. In ML this may describe scale or intensity. The context decides the interpretation.

```python
dot = v @ w
cosine = dot / (np.linalg.norm(v) * np.linalg.norm(w))
print('dot product:', dot)
print('cosine similarity:', cosine)
```

The dot product is one of the most important vector operations in all of ML. It combines corresponding coordinates and sums the results. That gives a measure related to both magnitude and alignment. Linear models, attention mechanisms, and similarity search all depend on this idea.

```python
a = np.array([1.0, 0.0])
b = np.array([0.0, 1.0])
print('orthogonal dot product:', a @ b)
```

Orthogonal vectors have zero dot product. Geometrically, they are at right angles. In data language, this can suggest independence of direction or lack of alignment. Keeping the geometry alive helps the formulas feel much less dry.

```python
points = np.array([[1, 1], [2, 1], [4, 5]])
query = np.array([2, 2])
distances = np.linalg.norm(points - query, axis=1)
print('distances to query:', distances)
print('nearest point:', points[np.argmin(distances)])
```

Nearest-neighbor reasoning depends directly on vector geometry. If you can picture points and distances, you already understand a large part of how similarity-based methods work.

## Final Takeaway
Vectors are the basic language of representation in ML. Length, direction, distance, and similarity are not separate topics; they are connected ways of understanding how numerical representations behave.
