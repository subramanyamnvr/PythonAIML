# Calculus as Change, Rate, and Slope
Calculus becomes far friendlier when it is introduced as a language of change rather than a wall of notation. In ML that matters because training is fundamentally about how small changes in parameters affect error.

A derivative measures how a function changes as its input changes. In plain language, it is a local slope or sensitivity. That idea is central in optimization, backpropagation, and error analysis.

```python
def f(x):
    return x**2 + 3*x + 1

def numerical_derivative(func, x, h=1e-5):
    return (func(x + h) - func(x - h)) / (2*h)

for point in [0, 1, 2]:
    print(point, numerical_derivative(f, point))
```

The numerical derivative above estimates slope by looking at nearby function values. Analytic calculus gives exact formulas, but numerical approximations help build intuition before the symbols settle in.

```python
xs = [0, 1, 2, 3]
ys = [f(x) for x in xs]
print(list(zip(xs, ys)))
```

If a function rises steeply, the derivative magnitude is large. If it flattens out, the derivative is small. If it changes from increasing to decreasing, you are near a turning point. Those intuitions are exactly the ones optimization uses.

## Final Takeaway
Calculus begins with the idea that functions do not just have values; they also have rates of change. In ML, learning uses those rates to decide how parameter updates should move.
