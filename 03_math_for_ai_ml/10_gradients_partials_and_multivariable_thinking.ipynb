# Gradients, Partials, and Multivariable Thinking
Most ML models do not depend on one variable. They depend on many parameters at once, which is why the derivative grows into partial derivatives and gradients. This notebook is about making that jump feel natural instead of abrupt.

A partial derivative measures how a function changes with respect to one variable while the others are held fixed. The gradient collects those partial derivatives into a vector that points in the direction of steepest increase.

```python
import numpy as np

def loss(w1, w2):
    return (w1 - 2)**2 + (w2 + 3)**2

def partial_w1(w1, w2, h=1e-5):
    return (loss(w1 + h, w2) - loss(w1 - h, w2)) / (2*h)

def partial_w2(w1, w2, h=1e-5):
    return (loss(w1, w2 + h) - loss(w1, w2 - h)) / (2*h)

point = (0.0, 0.0)
grad = np.array([partial_w1(*point), partial_w2(*point)])
print('gradient at', point, ':', grad)
```

If the gradient points toward steepest increase, then moving in the negative gradient direction moves toward steepest decrease locally. That single idea is the heart of gradient-based optimization.

```python
w = np.array([0.0, 0.0])
learning_rate = 0.1
for step in range(5):
    g = np.array([partial_w1(*w), partial_w2(*w)])
    w = w - learning_rate * g
    print('step', step, 'weights', w, 'loss', loss(*w))
```

Watching the parameters move numerically is often more helpful than staring at symbols alone. It shows that optimization is simply repeated local improvement guided by gradient information.

## Final Takeaway
Gradients turn multivariable change into a geometric object you can follow. Once you understand partial derivatives and the direction of steepest increase, gradient descent becomes much less mysterious.
