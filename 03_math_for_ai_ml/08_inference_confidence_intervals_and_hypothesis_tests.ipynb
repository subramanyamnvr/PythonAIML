# Inference, Confidence Intervals, and Hypothesis Tests
Inference is where mathematical confidence can become either genuinely useful or badly misleading. The key is to understand what a confidence interval or p-value is actually saying instead of treating it as a badge of certainty.

Inferential statistics helps us move from a sample to a claim about a population or effect. That movement is always conditional on assumptions, sample quality, and the question being asked.

```python
import math
import numpy as np
from scipy import stats

sample = np.array([52, 49, 51, 50, 53, 48, 54, 52, 51, 50], dtype=float)
mean = sample.mean()
std = sample.std(ddof=1)
n = len(sample)
se = std / math.sqrt(n)
critical = stats.t.ppf(0.975, df=n-1)
ci = (mean - critical * se, mean + critical * se)
print('mean:', mean)
print('95% CI:', ci)
```

A confidence interval gives a range produced by a procedure that would capture the true parameter a certain percentage of the time under repeated sampling assumptions. It is not a statement that the fixed true parameter is randomly moving around inside that specific interval.

```python
t_stat, p_value = stats.ttest_1samp(sample, popmean=50)
print('t-statistic:', t_stat)
print('p-value:', p_value)
```

A p-value tells you how surprising the observed statistic or something more extreme would be if the null hypothesis model were true. It is not the probability that the null hypothesis itself is true.

```python
interpretation_checks = [
    'Statistical significance is not the same as practical importance.',
    'A narrow interval often reflects more information or less noise.',
    'Bad sampling can invalidate polished-looking inference.',
    'Repeated testing without correction can create false discoveries.'
]
for item in interpretation_checks:
    print('-', item)
```

The best way to use inference in ML-adjacent work is with humility, careful assumptions, and attention to effect size and practical decision-making, not only threshold-based ritual.

## Final Takeaway
Confidence intervals and hypothesis tests are tools for disciplined uncertainty, not certainty machines. The deepest understanding comes from knowing what probability statement each tool does and does not justify.
