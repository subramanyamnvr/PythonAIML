# Probability from Events to Model Uncertainty
Probability becomes much less slippery when we stop treating it as a bag of formulas and start treating it as a disciplined way to talk about uncertainty. In ML that matters because data is noisy, labels are imperfect, and predictions are rarely all-or-nothing.

Probability gives a framework for talking about uncertain events, uncertain variables, and uncertain beliefs about outcomes. In practice, ML uses it to model noise, classify uncertainty, and compare competing hypotheses.

```python
from fractions import Fraction

coin = {'H': Fraction(1, 2), 'T': Fraction(1, 2)}
print(coin)
print('sum:', sum(coin.values()))
```

A probability distribution assigns nonnegative mass to outcomes and the total must sum to 1. That single rule already prevents many conceptual mistakes.

```python
def conditional(prob_ab, prob_b):
    return prob_ab / prob_b

print('P(A|B) example:', conditional(0.2, 0.5))
```

Conditional probability is central because information changes uncertainty. In ML, seeing features changes what we believe about labels, classes, or outcomes.

```python
def bayes(prior_a, likelihood_b_given_a, marginal_b):
    return prior_a * likelihood_b_given_a / marginal_b

print('Bayes example:', bayes(0.01, 0.95, 0.05))
```

Bayes' rule is not just an equation to memorize. It is a recipe for updating belief when new evidence arrives. That idea sits behind Naive Bayes, Bayesian modeling, and even everyday diagnostic reasoning.

```python
events = {
    'mutually_exclusive': 'cannot happen together',
    'independent': 'one event does not change the probability of the other',
}
for name, meaning in events.items():
    print(name, '->', meaning)
```

A common misunderstanding is mixing up mutual exclusivity and independence. Events that cannot happen together are generally not independent unless one of them has probability zero.

## Final Takeaway
Probability is the language of uncertainty. The most important habits are checking whether values define a valid distribution, understanding conditional probability, and seeing Bayes' rule as principled belief updating rather than a decorative formula.
