# Matrices, Linear Systems, and Transformations
A matrix is more than a rectangle of numbers. It can describe a linear transformation, a dataset, a system of equations, or a layer of learned parameters. Seeing those meanings together is what makes matrix operations start to feel natural.

If vectors are the language of single points or directions, matrices are the language of structured transformations and collections. Many ML operations become compact and efficient when written in matrix form.

```python
import numpy as np

A = np.array([[2.0, 1.0], [1.0, 3.0]])
x = np.array([4.0, 5.0])
print('A:\n', A)
print('x:', x)
print('Ax:', A @ x)
```

One way to read Ax is as a transformation applied to a vector. Another is as a weighted combination of the matrix columns. Both interpretations are useful, and both lead to the same numbers.

```python
b = np.array([9.0, 19.0])
solution = np.linalg.solve(A, b)
print('solution to Ax=b:', solution)
print('check:', A @ solution)
```

Solving Ax = b matters because many estimation and fitting problems reduce to finding parameters that satisfy or approximately satisfy such systems.

```python
X = np.array([[1, 2], [3, 4], [5, 6]])
print('rows as observations:\n', X)
print('column means:', X.mean(axis=0))
print('transpose:\n', X.T)
```

In data science and ML, a matrix often means rows are examples and columns are features. That interpretation lets you express preprocessing and model calculations cleanly.

```python
rotation_like = np.array([[0, -1], [1, 0]])
v = np.array([2.0, 1.0])
print('transformed vector:', rotation_like @ v)
```

Thinking of matrices as transformations is especially helpful for neural networks and dimensionality-reduction methods, where learned parameter matrices reshape data into new spaces.

## Final Takeaway
Matrices are the workhorses of ML because they can represent both data and transformations. Matrix multiplication, transposes, and linear systems are not separate mechanical skills; they are pieces of one representation-and-transformation story.
