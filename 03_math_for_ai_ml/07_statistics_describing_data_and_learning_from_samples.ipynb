# Statistics: Describing Data and Learning from Samples
Statistics asks how much we can learn from limited data without fooling ourselves. That balance between extracting signal and respecting uncertainty is exactly why it matters so much in machine learning and experimentation.

Probability often starts from a model and asks what data we should expect. Statistics often starts from data and asks what we can infer about the world or model behind it. The two topics are deeply connected but not identical.

```python
import numpy as np

data = np.array([12, 14, 14, 15, 18, 21, 21, 25], dtype=float)
print('mean:', data.mean())
print('median:', np.median(data))
print('variance:', data.var(ddof=1))
print('std:', data.std(ddof=1))
```

Mean and median answer different questions. The mean uses all values and is sensitive to outliers. The median tells you the middle location and is often more robust when data is skewed or noisy.

```python
q1, q2, q3 = np.percentile(data, [25, 50, 75])
print('quartiles:', q1, q2, q3)
print('IQR:', q3 - q1)
```

Spread is not one number with one correct formula. Range, variance, standard deviation, IQR, and MAD each describe variability differently. Good analysis chooses the summary that fits the data story.

```python
sample1 = np.array([10, 11, 12, 13, 12])
sample2 = np.array([20, 19, 18, 22, 21])
print('difference in means:', sample2.mean() - sample1.mean())
```

The central habit in statistics is humility. A sample mean is informative, but it is not the same thing as the true population mean. The gap between sample and population is where inference lives.

## Final Takeaway
Statistics is about extracting stable meaning from finite samples. Descriptive summaries tell you what the sample looks like, and inferential thinking reminds you that the sample is only a partial window into a larger process.
