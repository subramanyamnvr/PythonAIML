# Random Variables, Distributions, and Expectation
Once probability over named events feels comfortable, the next step is to package uncertain outcomes into variables and distributions. That move is what lets us talk about means, variance, likelihood, sampling, and model assumptions in a clean way.

A random variable maps outcomes to numerical values. The distribution tells us how likely different values are. This is what makes expectation and variance meaningful rather than just arithmetic exercises.

```python
import numpy as np

values = np.array([0, 1, 2])
probs = np.array([0.2, 0.5, 0.3])
expectation = np.sum(values * probs)
variance = np.sum(((values - expectation) ** 2) * probs)
print('expectation:', expectation)
print('variance:', variance)
```

Expectation is the weighted average under the distribution. Variance tells you how spread out the values are around that average. In ML, mean and variance show up everywhere from feature scaling to probabilistic modeling.

```python
rng = np.random.default_rng(0)
samples = rng.normal(loc=10, scale=2, size=10000)
print('sample mean:', samples.mean())
print('sample variance:', samples.var())
```

Sampling gives a practical bridge from theory to data. A theoretical distribution defines what could happen. A sample is one realized collection of values drawn from that distribution.

```python
bernoulli = rng.binomial(n=1, p=0.7, size=20)
print('bernoulli sample:', bernoulli)
print('mean approximation of p:', bernoulli.mean())
```

Bernoulli and binomial examples matter because they connect directly to classification, success-failure outcomes, and likelihood-based reasoning.

## Final Takeaway
Random variables and distributions let us represent uncertainty numerically. Expectation, variance, and sampling are the core tools that connect abstract probability to data you can actually observe and model.
