# Distance, Similarity, and Geometry in Model Space
A surprising amount of machine learning reduces to geometry: how far apart points are, which directions align, and what shape a cloud of data seems to occupy. This notebook brings those geometric intuitions together at the end of the math section.

Distance is not just one formula used for nearest neighbors. It is a modeling choice. Similarity is not just a convenience score. It changes what the system treats as close, relevant, or meaningful.

```python
import numpy as np

p = np.array([1.0, 2.0, 3.0])
q = np.array([2.0, 4.0, 6.0])
print('euclidean distance:', np.linalg.norm(p - q))
print('manhattan distance:', np.abs(p - q).sum())
print('cosine similarity:', (p @ q) / (np.linalg.norm(p) * np.linalg.norm(q)))
```

Euclidean distance measures straight-line separation. Manhattan distance adds coordinate-wise travel. Cosine similarity focuses on angle and ignores overall scale. Different tasks care about different notions of closeness.

```python
embeddings = np.array([[1.0, 0.0], [0.9, 0.1], [0.1, 0.9], [-1.0, 0.0]])
query = np.array([1.0, 0.0])
scores = embeddings @ query / (np.linalg.norm(embeddings, axis=1) * np.linalg.norm(query))
print('cosine similarities:', scores)
print('best match index:', np.argmax(scores))
```

Embedding-based retrieval, semantic search, clustering, and recommendation systems all depend on geometry in a representation space. The mathematics may look different in advanced systems, but the intuition is still about closeness and direction.

```python
design_questions = [
    'Should magnitude matter or only direction?',
    'Are features on compatible scales?',
    'Does the geometry match the task meaning?',
    'Will high dimensionality distort this metric intuition?'
]
for q in design_questions:
    print('-', q)
```

That last question about high dimensionality matters because intuition from two- or three-dimensional spaces does not always transfer cleanly to large feature spaces. Even so, geometric thinking remains one of the best anchors for understanding ML methods.

## Final Takeaway
Distance and similarity define the geometry of your problem. Choosing them well shapes retrieval, clustering, recommendation, and representation learning, which is why geometry remains central throughout AI and ML.
